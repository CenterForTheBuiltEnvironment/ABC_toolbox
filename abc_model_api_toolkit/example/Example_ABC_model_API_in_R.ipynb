# **Example code for ABC model API in R**

```r
# Clone the ABC_toolbox repository from GitHub
system("git clone https://github.com/CenterForTheBuiltEnvironment/ABC_toolbox.git")
```

```r
# Install the 'requests' Python package, which is required by the ABC model API client
system("python3 -m pip install requests")
```

```r
# Install and load the reticulate package for R-Python interoperability
# Please note that this takes 30 - 60 seconds.
install.packages("reticulate")
library(reticulate)
```

```r
# Add the ABC_toolbox directory to the Python path so we can import its modules
toolbox_path <- normalizePath("ABC_toolbox")
py_run_string(sprintf("import sys; sys.path.insert(0, '%s')", toolbox_path))
```

```r
# Import Python functions from the abc_model_api_toolkit package
get_results <- import("abc_model_api_toolkit.api_client")$get_abc_model_api_results
json_to_csv <- import("abc_model_api_toolkit.data_processor")$output_json_to_csv
```

```r
# Define the API endpoint
api_url <- "https://backend-384255928646.us-west1.run.app/abc"

# Set file paths (Please chenge the file paths here as you want)
input_json_path <- "/content/ABC_toolbox/abc_model_api_toolkit/example/example_input.json"
output_json_path <- "/content/ABC_toolbox/abc_model_api_toolkit/example/example_output.json"
output_csv_path <- "/content/ABC_toolbox/abc_model_api_toolkit/example/example_output.csv"
```

```r
# --- Call the ABC model API ---
results_path <- get_results(
  input_file = input_json_path,
  output_file = output_json_path,
  api_url = api_url
)

# --- Convert JSON output to CSV ---
json_to_csv(
  output_json_file_path = results_path,
  csv_file_path = output_csv_path
)

# --- Read the CSV result into R ---
example_csv <- read.csv(output_csv_path)

# Display the data
View(example_csv)
```
