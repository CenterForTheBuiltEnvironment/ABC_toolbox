```python
def run_abc(
    input_json_file,
    output_json_file,
    output_csv_file,
    api_url="https://backend-384255928646.us-west1.run.app/abc",
):
    """
    Function to run ABC model with your input json file.
    """
    call_abc_model_api(input_json_file, output_json_file, api_url)
    output_json_to_csv(output_json_file, output_csv_file)
```

```python
# Please change file paths as you want.
input_json_file="/content/ABC_toolbox/abc_model_api_toolkit/example/example_input.json"
output_json_file="/content/ABC_toolbox/abc_model_api_toolkit/example/example_output.csv"
output_csv_file="/content/ABC_toolbox/abc_model_api_toolkit/example/example_output.json"

run_abc(
    input_json_file=input_json_file,
    output_json_file=output_json_file,
    output_csv_file=output_csv_file,
)
```
